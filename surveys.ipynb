```python
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')

dirName = "/Users/rhl/TeX/Talks/LSST/Hiroshima-2015-02" if True else None
```

```python
area =     dict(DES=3.1, HSC=1.8, LSST=10.0, PanSTARRS=7, SDSS=7.0, TMT=40/60.0**2)
diameter = dict(DES=4.0, HSC=8.2, LSST=sqrt(8.4**2 - 5**2),  PanSTARRS=1.8, SDSS=sqrt(2.5**2 - 1.2**2), TMT=30)
diameter["PanSTARRS"] *= diameter["SDSS"]/2.5  # correct for secondary
```

```python
from matplotlib.patches import Circle

axes = plt.axes()

cen = (-2, 2, 6)

etendue = True

for i, survey in enumerate(sorted(area) + ["MOON"]):
    if survey == "MOON":
        if etendue:
            continue
        r = 0.25
        xy = (-0.5, 0)
    else:
        r = np.sqrt(area[survey]/np.pi)
        if etendue:
            r *= (diameter[survey]/diameter["LSST"])
        xy = (cen[i//2], cen[i%2])
    circ = Circle(xy=xy, radius=r, alpha=0.4, color='black' if survey == "MOON" else None)
    axes.add_artist(circ)
        
    plt.text(xy[0] + 0.8*r, xy[1] + 0.8*r, survey)

    if survey == "LSST":
        if etendue:
            r /= sqrt(21)    # ComCam -- 1 raft out of 21
            circ = Circle(xy=xy, radius=r, alpha=0.4)
            axes.add_artist(circ)
    elif survey == "PanSTARRS":
        r *= sqrt(2)
        circ = Circle(xy=xy, radius=r, alpha=0.4)
        axes.add_artist(circ)       
    elif survey == "SDSS":
        r = np.sqrt(30*(2048*0.396/3600)**2/np.pi)   # area of CCDs
        if etendue:
            r *= diameter[survey]/diameter["LSST"]
        
        for fac in (1, 1/sqrt(5)):                   # area per band
            circ = Circle(xy=xy, radius=r*fac, alpha=0.4)
            axes.add_artist(circ)
    
plt.xlabel("" if etendue else "degrees")
plt.title("Etendue" if etendue else "Field of View")

plt.xlim(-4.0, 8)
plt.ylim(-4, 4.5)
axes.set_aspect('equal')

if dirName:
    plt.savefig(os.path.join(dirName, "etendue.pdf" if etendue else "focalPlanes.pdf"))
```
