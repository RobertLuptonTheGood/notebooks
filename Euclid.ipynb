```python
%matplotlib inline
import os
import numpy as np
import matplotlib.pyplot as plt

import lsst.afw.image as afwImage
import lsst.afw.table as afwTable
```

```python
inDir = os.path.expanduser("~/Euclid/realization_2")
catTemplate = "cat-%s.fits"
expTemplate = "calexp-%s.fits"

cat = {}
calib = {}
exposure = {}
filters = "gri"
for f in filters:
    cat[f] = afwTable.SourceCatalog.readFits(os.path.join(inDir, catTemplate % f))
    calexpFile = os.path.join(inDir, expTemplate % f)
    if False:
        exposure[f] = afwImage.ExposureF.readFits(calexpFile)
        calib[f] = exposure[f].getCalib()
    else:
        md = afwImage.readMetadata(calexpFile)
        calib[f] = afwImage.Calib(md)
```

```python
colors = dict(g='green', r='red', i='magenta')
for f in filters:
    plt.scatter(cat[f].getX(), cat[f].getY(), marker='.', color=colors[f], edgecolors="none")
```

```python
from lsst.analysis.utils import zipMatchList, getFieldSuffixes

matchRadius = 1				# pixels
filters = ['g', 'r', 'i']
m = zipMatchList(afwTable.matchXy(cat[filters[0]], cat[filters[1]], matchRadius))
for f in filters[2:]:
    m = zipMatchList(afwTable.matchXy(m, cat[f], matchRadius))
    
suffixes = getFieldSuffixes(filters)

def getMags(magType, filters):
    if magType in ["cmodel",]:
        flux = "%s.flux" % magType
    else:
        flux = 'flux.%s' % magType
    

    mags = []
    for f in filters:
        calib[f].setThrowOnNegativeFlux(False)
        mags.append(calib[f].getMagnitude(m.get(flux + suffixes[f])) + 23.9)
        
    return mags
```

```python
def plotCM(f1, f2, magType, xlim=(0, 3), ylim=(26, 20)):
    """Plot a colour-magnitude diagram f1 - f2 v. f1"""

    m1, m2 = getMags(magType, [f1, f2])
    plt.scatter(m1 - m2, m2, marker='.', color=colors[f], edgecolors="none")
    plt.xlim(xlim)
    plt.ylim(ylim)
    plt.xlabel('%s - %s' % (f1, f2))
    plt.ylabel(f2)
    plt.title(magType)
    
def plotCC(f1, f2, f3, magType, good=None, xlim=(0, 3), ylim=(-1.5, 1.5)):
    """Plot a colour-colour diagram f1 - f2 v. f2 - f3"""

    m1, m2, m3 = getMags(magType, [f1, f2, f3])
    x, y = m1 - m2, m2 - m3
    if good is not None:
        x,y  = x[good], y[good]
    plt.scatter(m1 - m2, m2 - m3, marker='.', color=colors[f], edgecolors="none")
    plt.xlim(xlim)
    plt.ylim(ylim)
    plt.xlabel('%s - %s' % (f1, f2))
    plt.ylabel('%s - %s' % (f2, f3))
    plt.title(magType)
```

```python
good = getMags('cmodel', "r") < 24
```

```python
plotCM('g', 'r', 'cmodel')
```

```python
plotCC('g', 'r', 'i', 'cmodel', good)
```

```python
plotCC('g', 'r', 'i', 'sinc', good)
```
