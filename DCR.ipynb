```python
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

which are based on Filippenko 1982.  

Based on 0.6" seeing, but I don't know how they
converted a FWHM to an rms size, which is what matters here.

```python
zenith = np.linspace(0, 90, 100)
FWHM = 0.6                      # assumed seeing, arcsec
dcrSigma = FWHM/(2*sqrt(2*log(2))) # converted to rms assuming a Gaussian
pixelSize = 0.200               # pixel size in arcsec

dcr = {}
dcr["g"] = 0.3*(zenith/65)**2
dcr["r"] = 0.05*(zenith/60)**2 + 0.2*(zenith/83)**6
dcr["i"] = 0.01*(zenith/60)**2 + 0.2*(zenith/83)**8

colors = dict(g="green", r="red", i="magenta")

for b in ("g", "r", "i",):
    plt.plot(zenith, dcr[b], color=colors[b], label=b)

plt.legend(loc="best")

plt.xlabel("Zenith distance (degrees)")
plt.ylabel("PSF ellipticity $\equiv 1 - b^2/a^2$")

plt.xlim(-2, 85)
ylim = plt.ylim(-0.03, 0.3)
```

If the DCR results in a uniform smear along a vector of length $\delta$ then if the rms size of the seeing is $\sigma$, then (because rms sizes add and $\int_{-1/2}^{1/2} x^2\,dx = 1/12$) we have

$$
e \equiv \frac{a^2 - b^2}{a^2} = \frac{\sigma^2 + \delta^2/12 - \sigma^2}{\sigma^2} = \frac{1}{12}\left(\frac{\delta}{\sigma}\right)^2
$$
so
$$
\delta = \sqrt{12 e}\,\sigma
$$

```python
for b in ("g", "r", "i",):
    delta = np.sqrt(12*dcr[b])*dcrSigma
    delta /= pixelSize
    plt.plot(zenith, delta, color=colors[b], label=b)

plt.legend(loc="best")

plt.xlabel("Zenith distance (degrees)")
plt.ylabel("Smear (pixels)")

plt.xlim(-2, 60)
ylim = plt.ylim(-0.03, 1.9)
```

Calculate the profiles of (Gaussian N(0, sigma^2)) PSFs smeared from (-delta/2 ... delta/2)

```python
def smear(x, delta, sigma=1.0):
    """Smear an N(0, sigma^2) PSF with a straight line (-delta/2, delta/2)"""
    if delta == 0:
        return 1/sqrt(2*np.pi*sigma**2)*np.exp(-(x/sigma)**2/2)
    
    import scipy.special as special

    return 0.5*(special.erf(1/sqrt(2)*((x + 0.5*delta)/sigma)) - 
                special.erf(1/sqrt(2)*((x - 0.5*delta)/sigma)))/delta

def smearBruteForce(x, delta, sigma=1):
    """Do the smear by brute force, summing a set of shifted PSFs (see also smear())"""
    if delta == 0:
        return 1/sqrt(2*np.pi*sigma**2)*np.exp(-(x/sigma)**2/2)
    
    ans = np.zeros_like(x)
    for d in np.linspace(-delta/2, delta/2, 101):
        ans += np.exp(-((x - d)/sigma)**2/2)
    ans /= np.sum(ans)*(x[1] - x[0])

    return ans
```

Calculate approximate smeared profiles by assuming that it's a Gaussian with a suitable width.

```python
def smearApproxSigma(delta, sigma=1):
    """Return the s.d. needed to approximate the smear as a Gaussian
    
    Works reasonably well out to delta ~ 2.0
    """
    return np.sqrt(sigma**2 + float(delta)**2/12)

def smearApprox(x, delta, sigma=1):
    """Approximate the smear as resulting in a Gaussian (see also smear())"""
    return smear(x, 0.0, smearApproxSigma(delta, sigma=sigma))

def largestResidualApprox(delta, sigma=1):
    """Return the largest |value| of the residual given by subtracting an unsmeared PSF from a trail"""
    return 1/sqrt(2*np.pi)*(1/smearApproxSigma(delta, sigma) - 1/sigma)
```

```python
x = np.linspace(-5, 10, 401)
dx = x[1] - x[0]

delta = float(2)
decenter = 0.20

psf0 = smear(np.array([0]), 0)
psf = smear(x, 0)/psf0
smeared = smear(x - decenter, delta)/psf0
residuals = smeared - psf

plt.plot(x, psf, label="PSF")
plt.plot(x, smeared, label="PSF smeared by %g" % delta)
plt.plot(x, residuals, label="smeared - psf")
if False:
    plt.plot(x, smearApprox(x - decenter, delta)/psf0 - smeared, label="Approx - smeared")

plt.legend(loc='best')

plt.axvline(decenter, ls=':')
plt.axvspan(-0.5*delta, 0.5*delta, facecolor='g', alpha=0.2)

plt.xlabel("x (pixels)")
plt.ylabel("Intensity")

ylim = (min([-0.01, np.min(residuals)]),
        max([np.max(psf), np.max(smeared)]))
ylim = plt.ylim(ylim + 0.05*np.array((-0.5, 1))*(ylim[1] - ylim[0]))
```

We can now use this way of thinking about DCR to plot the maximum residual (assuming that the astrometry is done such that the smeared image is aligned with the unsmeared image so that the offsets in the centroids is decenter).

```python
psf0 = smear(np.array([0]), 0)
psf = smear(x, 0)

for b in ("g", "r", "i",):
    delta = np.sqrt(12*dcr[b])*dcrSigma
    delta /= pixelSize
    
    residual = np.empty_like(delta)
    for decenter, lt in [(0.0, '-'), (0.05, '--'), (0.10, ':')]:
        for i, s in enumerate(delta):
            residual[i] = 100*np.max(np.abs((smear(x + decenter*delta[i], delta[i]) - psf)/psf0))
        plt.plot(zenith, residual, color=colors[b], linestyle=lt, 
                 label=b if decenter == 0.0 else None)

plt.legend(loc="best")

plt.xlabel("Zenith distance (degrees)")
plt.ylabel("Amplitude of residual (%)")
plt.title("""\
Residuals subtracting zenith from DCR image
solid: centred; dotted: offset by %g$\delta$""" %
          (decenter))

plt.xlim(-2, 60)
ylim = plt.ylim(-1.5, 15)
```
