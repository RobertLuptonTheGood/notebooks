```python
class LENS: pass
class DETECTOR: pass
class MIRROR: pass
class SPACER: pass
class STOP: pass

class Optics(object):
    """An optical system;  add elements then ray trace it"""
    
    def __init__(self, name="", unfold=False, label=True):
        """Create an optical system"""
        self.name = name        
        self._elements = []
        self._location = 0.0
        self._direction = +1       # propagating in the +x direction
        self._unfold = unfold      # onfold the optical system treating mirrors as lenses?
        self._label = label        # label optical elements
        self.add(0)                # We have to start somewhere
        
    def add(self, param, what=SPACER, name=""):
        if what == SPACER:
            self._location += self._direction*param

        el = [what, self._location, name]
        if what in (LENS, MIRROR):
            f, rad = param
            f *= 2*rad
            extra = rad       
            if what == MIRROR:
                el.append(self.mirror(f))
                if not self._unfold:
                    self._direction *= -1
            else:
                el.append(self.lens(f))
        elif what == SPACER:
            extra = np.nan
            el.append(self.propagate(param))
        elif what in (DETECTOR, STOP):
            extra = param
            el.append(self.stop())
            
        el.append(extra)
        self._elements.append(el)
        
    def M(self, n=None):
        if n is None:
            n = len(self._elements) - 1
            
        M = np.matrix([[1, 0], [0, 1]])
        for _what, _loc, _name, _dM, _extra in self._elements[0:n+1]:
            M = _dM*M
            
        return M

    def lens(self, f):
        """A lens of focal length f"""
        return np.matrix([[1, 0], [-1/float(f), 1]])

    def mirror(self, f):
        """A mirror of focal length f"""
        return np.matrix([[1, 0], [-1/float(f), 1]])

    def propagate(self, d):
        """Propagate a distance d in free space"""
        return np.matrix([[1, float(d)], [0, 1]])

    def stop(self):
        """Don't modify rays at all"""
        return np.matrix([[1, 0], [0, 1]])

    def draw(self, y, theta=[0]):
        x = np.empty(len(self._elements))
        z = np.empty(len(self._elements))
    
        x0 = 0         # position of first real optical element
        for el in self._elements:
            _what, _loc = el[0:2]
            _name = el[2]
            if _what != SPACER:
                break
            x0 = _loc
        #
        # Draw the optics
        #
        titlePM = +1  # +-1: draw at top or bottom?
        for _what, _loc, _name, _dM, _extra in self._elements:
            _loc -= x0
            if self._label and _name:
                plt.text(_loc, titlePM*1.2*_extra, _name,
                                verticalalignment='center')
                titlePM *= -1
            
            if _what in (DETECTOR, LENS, MIRROR):
                plt.plot([_loc, _loc], [-_extra, _extra], 
                         color={DETECTOR:'black', LENS:'red', MIRROR:'blue'}[_what])
            elif _what == STOP:
                big = 1.2*max([abs(_[4]) for _ in self._elements])

                if _extra > 0:
                    r1, r2 = _extra, 4
                else:
                    r1, r2 = 0, _extra

                for pm in [-1, 1]:
                    plt.plot([_loc, _loc], pm*np.array([r1, r2]), color='green')
        #
        # Trace the rays
        #
        colors = ["black", "cyan", "magenta", "yellow", "brown"]
        for yy in y:
            for j, tt in enumerate(theta):
                for i in range(len(self._elements)):
                    _what, _loc, _name, _dM, _extra = self._elements[i]
                
                    x[i] = _loc
                    z[i] = (tel.M(i)*np.matrix([[yy], [tt]]))[0]
                
                    if _what in (DETECTOR, LENS, MIRROR, STOP):
                        if (_extra > 0 and abs(z[i]) > _extra) or \
                           (_extra < 0 and abs(z[i]) < abs(_extra)):
                            z[i+1:len(z)] = np.nan
                            break
                
                plt.plot(x - x0, z, '-', alpha=0.1, color=colors[j%len(colors)])
            
        plt.title(self.name)
        #
        # Set better limits in x/y
        #
        if True:
            for zlim in (plt.xlim, plt.ylim):
                lim = zlim();  dx = lim[1] - lim[0]; zlim(lim[0] - 0.05*dx, lim[1] + 0.04*dx)
```

```python
# Mn: mirrors; FP: focal plane
diameter = dict(M1=[5.12, 8.36],  M2=[1.80, 3.42],  M3=[1.22, 5.02],  FP=[0, 0.64])   # ID/OD (m)
fratio   = dict(M1=1.14, M2=-1.05, M3=1.19, FP=np.inf)
spacing  = dict(M1=0,    M2=6.39, M3=6.39, FP=4.66)

diameter["CORR"] = [0, 1.55]
fratio["CORR"] = 6
spacing["CORR"] = spacing["FP"] - 1.03;  spacing["FP"] -= spacing["CORR"]

# approximate corrections for thickness of mirrors
spacing["M2"] -= 0.20

if True:                    # gets the size of the beam right at the corrector, and makes it slower
    fratio["M3"] *= 0.72 # 0.85
    fratio["M2"] *= 0.85

# stops aren't quite right in my model
if False:                      # Not needed if only going out to 1.5degrees
    diameter["M3"][0] = 0
    diameter["CORR"][1] += 0.05
```

```python
tel = Optics("LSST", unfold=False, label=True)

tel.add(7)

for M in ("M1", "M2", "M3", "CORR", "FP"):
    tel.add(spacing[M])
    if M == "FP":
        tel.add(0.5*diameter[M][1], DETECTOR, name="FP")
    else:
        tel.add((fratio[M], 0.5*diameter[M][1]), MIRROR if M[0] == 'M' else LENS, name=M)
        tel.add(-0.5*diameter[M][0], STOP)

y = np.linspace(-0.5*diameter["M1"][1], 0.5*diameter["M1"][1], 50)
y = y[np.where(np.abs(y) >= 0.5*diameter["M3"][1])]
thetaMax = np.radians(1.5)            # really 1.75
tel.draw(y, np.linspace(-thetaMax, thetaMax, 3))

if False:
    plt.axes().set_aspect('equal')

#plt.xlim(-5, -3); _ = plt.ylim(-1.5, 1.5)
#print ["%s %g" % (_[2], _[1] - 7.3) for _ in tel._elements if _[2]]
```
