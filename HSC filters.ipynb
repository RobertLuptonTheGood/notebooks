```python
filterColor = dict(
    g  = 'green',
    r  = 'red',
    r1 = 'orange',
    i  = 'magenta',
    z  = 'black',
    y  = 'brown',
)
```

```python
def makeEvenSplineInterpolator(x, y):
    npt = len(x)

    vecs = [x, y]
    for i, v in enumerate(vecs):
        v_o = np.empty(2*npt - 1)
        vecs[i] = v_o

        v_o[0:npt] = v[::-1]
        if i == 0:                      # radius
            v_o[0:npt] *= -1
        v_o[npt:] =   v[1:]

    return scipy.interpolate.interp1d(vecs[0], vecs[1], kind='cubic')

def plotRadialFilterData(what, filters="grizy", doSpline=False):
    options = ["EW", "lambda"]
    if what not in options:
        raise RuntimeError("Please choose one of %s" % " ".join(options))

    plt.clf()

    ymin, ymax = None, None
    for b in filters:
        filterR = filterData[b]["radius"]
        filterR = filterR/15e-3         # convert to pixels

        if what == "EW":
            y = filterData[b]["EW"]
            y = y/y[0]
        elif what == "lambda":
            y = filterData[b]["lambda_bar"]
            n = len(y)
            xvec = np.empty(2*n);  yvec = np.empty_like(xvec)
            xvec[0:n] = filterR;  xvec[n:] = filterR[::-1]

            if ymin is None:
                ymin, ymax = 1e30, -1e30

            for i in (10, 50, 80):
                alpha = 0.1 + 5e-3*i

                yvec[0:n] = filterData[b]["on%d" % i]
                yvec[n:] =  filterData[b]["off%d" % i][::-1]

                p = plt.Polygon(zip(xvec, yvec), color=filterColor[b], alpha=alpha)
                plt.axes().add_artist(p)

                ymin, ymax = (min(ymin, np.min(yvec)), max(ymax, np.max(yvec)))
        else:
            raiseRuntimeError("You can't get here")
            
        plt.plot(filterR, y, color=filterColor[b], label=b)
        plt.plot(filterR, y, "o", color=filterColor[b])
        if scipy and doSpline:
            if what == "EW":
                interpolator = makeEvenSplineInterpolator(filterR, y)
                r = np.linspace(0, filterR.max(), 100)
                plt.plot(r, interpolator(r), "--", color=filterColor[b])

    plt.xlabel("Radius (pixels)")
    if what == "EW":
        plt.ylabel("Equivalent Width (relative to R==0)")
    elif what == "lambda":
        plt.ylabel(r"$\lambda$ (nm)")
        d = ymax - ymin
        plt.ylim(ymin - 0.1*d, ymax + 0.1*d)
    else:
        raiseRuntimeError("You can't get here")

    legend = plt.legend(loc="best")
    if legend:
        legend.draggable(True)
    plt.show()
    
def plotFilterBandpass(filters="grizy", index=None):
    """Plot the filter bandpasses"""
    plt.clf()
    
    for b in filters:
        filterR = filterData[b]["radius"]
        filterR = filterR/15e-3         # convert to pixels
        
        trans = filterData[b]["Tavg"]
        Tavg = np.sum(trans*filterR/sum(filterR)) if index is None else trans[index]

        yvec = np.array([0, 10, 50, 80, Tavg, Tavg, 80, 50, 10, 0])
        xvec = np.empty_like(yvec)

        for i, key in enumerate((None, "on10", "on50", "on80", None, None, 
                                 "off80", "off50", "off10", None)):
            if not key:
                continue

            trans = filterData[b][key]
            xvec[i] = np.sum(trans*filterR/sum(filterR)) if index is None else trans[index]
        # Linear extrapolation to find 0 and Tavg points
        xvec[0] = 2*xvec[1] - xvec[2]
        xvec[4] = 2*xvec[3] - xvec[2]
        xvec[5] = 2*xvec[6] - xvec[7]
        xvec[-1] = 2*xvec[-2] - xvec[-3]

        plt.text(np.sum(xvec*yvec/sum(yvec)), Tavg - 10, b, color=filterColor[b])
        
        plt.plot(xvec, yvec, color=filterColor[b])
        plt.plot(xvec, yvec, "o", color=filterColor[b])

    plt.ylim(-5, 105)
    plt.xlabel("$\lambda$ (nm)")
    plt.ylabel("Throughput, %")

    plt.title("Filter transmission (%s)" % 
              ("mean radius" if index is None else ("at %.0fpixels" % filterR[index])))

    plt.show()
```

```python
plotRadialFilterData("EW", filters="grizy", doSpline=True)
```

```python
plotRadialFilterData("lambda", filters="grizy")
```

```python
plotFilterBandpass(filters="grizy")
```
