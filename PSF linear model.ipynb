```python
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def makePsf(x, sigma1=1.0, b=0.0, sigma_ratio=2, xc=0):
    I = np.exp(-0.5*((x - xc)/sigma1)**2) + b*np.exp(-0.5*((x - xc)/(sigma_ratio*sigma1))**2)
    I /= np.sum(I)*(x[1] - x[0])
    
    return I
```

```python
x = 5*np.linspace(-1, 1, 30)

plt.plot(x, makePsf(x))
plt.plot(x, makePsf(x, b=0.1))
plt.plot(x, makePsf(x, b=0.2))

plt.show()
```
