```python
calexps[patch] = butler.get('deepCoadd', immediate=True, tract=tract, patch=patch, filter=filter)

ccds = calexps[patch].getInfo().getCoaddInputs().ccds

srcs[patch] = butler.get('deepCoadd_src', immediate=True, tract=tract, patch=patch, filter=filter)
```

```python
visits = [1266, 1282, 1286, 1292, 1304, 1306, 1312, 1314, 1322]
visits = [v for v in visits if v not in (1304, 1314,)]

for i, visit in enumerate(visits):
    key = "%s-%d" % (patch, visit)
    
    if not key in calexps:
        calexps[key] = butler.get("deepCoadd_tempExp", tract=tract, patch=patch, filter=filter, visit=visit)
    
    for v, ccd in zip(ccds.get('visit'), ccds.get('ccd')):
        if v != visit:
            continue
            
        _src = butler.get("src", visit=int(v), ccd=int(ccd))
        
        if True:                # Use meas_mosaic astrometry
            _wcs = butler.get("wcs", tract=tract, visit=int(v), ccd=int(ccd)).getWcs()
        
            centroidKey = _src.getSchema().find(_src.getTable().getCentroidDefinition()).getKey()
            coordKey =    _src.getSchema().find("coord").getKey()

            for s in _src:
                s.set(coordKey, _wcs.pixelToSky(s.get(centroidKey)))

        if key not in srcs:
            srcs[key] = _src
        else:
            srcs[key].extend(_src, True)
        del _src
```

Calculate the vectors from coadd position to the mean of the input positions

```python
matchRadius = 1*afwGeom.arcseconds
nmatchMin = len(srcs) - 1          # don't need one for the patch catalogue itself
nmatchMin -= 2

matchedIds = {}
for key in srcs:
    if key == patch:
        continue
    #print key, len(srcs[key])
    
    for s1, s2, d in afwTable.matchRaDec(srcs[patch], srcs[key], matchRadius):
        id1 = s1.getId()
        if id1 not in matchedIds:
            matchedIds[id1] = [s1]
        matchedIds[id1].append(s2)

id = np.array(sorted([k for k, v in matchedIds.items() if len(v) > nmatchMin]), dtype=np.int64)
good = np.empty_like(id, dtype=bool)
x = np.empty_like(id, dtype=float)
y = np.empty_like(x)
dx = np.empty_like(x)
dy = np.empty_like(dx)

for i, _id in enumerate(id):
    matches = matchedIds[_id]
    s1 = matches[0]
    
    good[i] = not (s1.get("flags.pixel.edge") or s1.get("flags.pixel.interpolated.center"))

    x[i] = s1.getCoord()[0].asDegrees()
    y[i] = s1.getCoord()[1].asDegrees()
    
    dx[i] = np.mean([(m.getCoord()[0] - s1.getCoord()[0]).asArcseconds() for m in matches[1:]])
    dy[i] = np.mean([(m.getCoord()[1] - s1.getCoord()[1]).asArcseconds() for m in matches[1:]])
```

```python
pivot='middle'
scale=5                     # larger => smaller arrows
width=5e-4
labelLen = 0.10
maxLen = 0.10
clipLen = 1.2*maxLen

use = np.logical_and(good, np.hypot(dx, dy) < clipLen)

Q = plt.quiver(x[use], y[use], dx[use], dy[use], np.hypot(dx, dy)[use], 
               norm=pyplot.Normalize(0, maxLen),
               #color=plt.cm.jet, 
               units='xy', scale_units='xy',
               width=width, pivot=pivot, scale=scale, edgecolor="none",
               headlength=1.5, headwidth=2.5, headaxislength=2.25,
               )
cb = plt.colorbar()
cb.ax.get_yaxis().labelpad = 15
cb.set_label('arcsec (clipped at %g)' % (clipLen), rotation=270)

plt.quiverkey(Q, 0.15, 0.05, labelLen, "%g arcsec" % (labelLen), coordinates='axes',
              color='black', labelpos='E')

xlim = (214.75, 214.51)
ylim = (0.831, 1.030)
ylim = np.mean(ylim) + np.array([-0.5, 0.5])*np.abs(xlim[1] - xlim[0])*np.cos(np.mean(np.deg2rad(y)))

plt.xlim(xlim)
plt.ylim(ylim)
plt.axes().set_aspect("equal")

plt.xlabel("Ra (degrees)")
plt.ylabel("Dec (degrees)")
plt.title("Rerun %s tract %d patch %s %s" % (rerun, tract, patch, filter[-1:].lower()))
#             ", ".join([str(v) for v in visits])

if False:
    for axline in (plt.axhline, plt.axvline):
        axline(0.0, color='blue', ls=':')

plt.savefig('offsets-%s.pdf' % rerun)
```
