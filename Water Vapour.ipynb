Extract equivalent width of water feature from modtran transmission
spectrum that has been degraded. Data files from Patrick Ingraham. 

C. Stubbs Sept 10 2015

```python
import os
import numpy as np
from scipy import interpolate, signal
import pyfits

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def readModtrans(seeing, smoothfactor=None):
    """Read a seeing file
    seeing:  value of seeing (which defines the file to read)
    smoothfactor: width of boxcar to smooth data (or None)
    """
    infile = 'Atm_binned_%.1farcsec_seeing.fits' % seeing
    inDir = os.path.expanduser("~/LSST/Calibration")
    tbl = pyfits.open(os.path.join(inDir, infile))
    
    data = tbl[0].data
    lam, transmission = data
    lam *= 0.1                # convert from \AA to nm
    
    if smoothfactor:
        filt = signal.get_window('boxcar', smoothfactor)
        filt /= np.sum(filt)
        transmission = signal.convolve(transmission, filt, 'same')

    # Throw away the region without significant water absorption
    if True:
        keep = np.logical_and(lam >= 864.0, lam <= 1010.0)
    else:
        keep = np.logical_and(lam >= 840.0, lam <= 1110.0)

    lam = lam[keep]
    transmission = transmission[keep]        

    return lam, transmission

seeing = 0.5
lam, transmission = readModtrans(seeing)
```

```python
def resample(x, y, binsize):
    """Bin x and y into bins of width binsize"""
    res = []
    for z in (x, y):
        if len(z)%binsize != 0:
            z = z[0:binsize*(len(z)//binsize)]
        shape = (len(z)//binsize, binsize)
        z = np.mean(z.reshape(shape), axis=1)

        res.append(z)
        
    return res
```

```python
def estimateContinuum(lam, transmission, method='linear'):
    """Fit the continuum near the water feature at 940 nm
    """
    # Good unattenuated wavelengths are 864.0-873.0 and 993.5-1000.0 nm
    sel = np.logical_or(lam <= 873.0, lam >= 993.5)

    if method == 'linear':
        interp = np.poly1d(np.polyfit(lam[sel], transmission[sel], 1))
    elif method == 'spline':
        interp = interpolate.InterpolatedUnivariateSpline(lam[sel],
                                                         transmission[sel])
    else:
        raise RuntimeError("Unknown method %s" % method)
    
    return interp(lam)

continuum = estimateContinuum(lam, transmission)
```

```python
def calculateEW(lam, transmission, continuum):
    return np.trapz(1 - transmission/continuum, lam)

EW = calculateEW(lam, transmission, continuum)
```

```python
plt.plot(lam, transmission, label='MODTRANS')
plt.plot(lam, continuum, label='linear')
if False:
    plt.plot(lam, estimateContinuum(lam, transmission, 'spline'),
             label='spline')

plt.xlim(None, 1005)
plt.xlabel(r"$\lambda$ (nm)")
plt.ylabel("Transmission")
plt.title('%g arcsec' % seeing)
_ = plt.legend(loc="best")
```

```python
def plotTransmission(lam, transmission, continuum, binsize=None, alpha=1.0):
    plt.plot(lam, transmission/continuum, label='raw', alpha=alpha)
    if binsize:
        blam, btransmission = resample(lam, transmission, binsize=binsize)
        bcontinuum = estimateContinuum(blam, btransmission)
        plt.plot(blam, btransmission/bcontinuum, 'o',
                 label='binned %d' % binsize, alpha=alpha)
        plt.legend(loc='best')
        
    plt.axhline(1.0, ls=':', color='black')
    plt.xlabel(r"$\lambda (nm)$")
    plt.ylabel("Fractional Transmission")
    plt.title(r"%g asec  EW %.2fnm" % (seeing,
                                calculateEW(lam, transmission, continuum)))
    
plotTransmission(lam, transmission, continuum, binsize=16)
```

```python
binsize = 4
nSample = 1000
SN0 = 15   # per nm

SN = SN0*sqrt(lam[1] - lam[0])  # per pixel

blam, btransmission = resample(lam, transmission, binsize=binsize)
bcontinuum = estimateContinuum(blam, btransmission)

ylim = None
EW = np.empty(nSample)
for fitContinuum in (True, False):
    np.random.seed(666)
    for i in range(nSample):
        trans = np.random.normal(transmission, 1/float(SN))
        trans = resample(lam, trans, binsize=binsize)[1]
        
        cont = estimateContinuum(blam, 
                                 trans if fitContinuum else btransmission)
        if True:
            # Fit a model by ML  absorption = A*(modtrans absorption)
            d, m = 1 - trans/cont, 1 - btransmission/bcontinuum # data, model
            A = np.sum(m*d)/np.sum(m**2)
            EW[i] = np.trapz(A*m, blam)

            plotTransmission(blam, 1 - A*m, 1, alpha=1)
        else:
            # Naive estimate
            EW[i] = calculateEW(blam, trans, cont)
    
        plotTransmission(blam, trans, cont, 
                         alpha=max(0.05, 4.0/(max([4, nSample]))))
        
    mEW, dEW = np.mean(EW), np.std(EW)

    title =  r"%g asec  EW %.2f $\pm$ %.2fnm" % \
              (seeing, mEW, dEW)
    title += r" %.1f [S/N per nm %.1f]" % \
              (mEW/dEW, SN0)
    if fitContinuum:
        title += " (fit continuum)"
    plt.title(title)
    ylim = plt.ylim(ylim)
    plt.show()
    
plt.hist(EW, 20)
plt.xlabel("EW")
_ = plt.axvline(mEW, color='black', ls=':')
```

Calculate the the theoretical S/N

```python
if True:
    l = blam
    t, c = btransmission, bcontinuum
else:
    l = lam
    t, c = transmission, continuum

dlambda = l[1] - l[0]
m = 1 - t/c           # model of absorption
    
sigma = 1/(SN0*sqrt(dlambda))
nEff = np.sum(m)**2/np.sum(m**2)
    
EW  = dlambda*np.sum(m)
dEW = dlambda*sigma*sqrt(nEff)

eSN = SN0*sqrt(np.trapz(m**2, l))

print "EW = %.2f +- %.2f nm  (S/N = %.1f)  S/N per nm: %g" % \
                (EW, dEW, eSN, SN0)
```

```python
binsize = np.array([1, 2, 3, 4, 8, 12, 16, 24,], dtype=float)
SN = np.empty_like(binsize)
SNnaive = np.empty_like(binsize)
resolution = np.empty_like(binsize)

for i, bin in enumerate(binsize):
    blam, btransmission = resample(lam, transmission, binsize=bin)
    bcontinuum = estimateContinuum(blam, btransmission)

    m = 1 - btransmission/bcontinuum           # model of absorption
    sigma = 1/(SN0*sqrt(blam[1] - blam[0]))

    if False:
        SN[i] = SN0*sqrt(np.trapz(m**2, blam))
    else:
        nEff = np.sum(m)**2/np.sum(m**2)
        SN[i] = np.sum(m)/(sigma*sqrt(nEff))
            
    SNnaive[i] = np.sum(m)/(sigma*sqrt(len(m)))

    resolution[i] = 0.5*(blam[1] + blam[0])/(blam[1] - blam[0])

if True:
    plt.plot(resolution, SN, label="MLE")
    plt.plot(resolution, SNnaive, label=r"Na$\"\i$ve")
    plt.legend(loc='best')
else:
    plt.errorbar(resolution, SN, SN/sqrt(2*nSample))

plt.xscale('log')
plt.xlim(40, 2e3)

plt.xlabel("Resolution")
plt.ylabel("EW's S/N")
_ = plt.title("S/N per nm: %g" % SN0)

print "S/N in EW:  MLE %.1f  Naive %.1f" % (SN[0], SNnaive[0])
```
