```python
%matplotlib inline
import os
import numpy as np
import matplotlib.pyplot as plt

import lsst.daf.persistence as dafPersist
import lsst.afw.geom as afwGeom
import lsst.afw.image as afwImage
import lsst.afw.table as afwTable
```

```python
def findSource(sourceSet, x, y, radius=2):
    sch = afwTable.SourceTable.makeMinimalSchema()

    centroidName = "foo"
    sch.addField(afwTable.Field["PointD"](centroidName, ""))
    oneObjCatalog = afwTable.SourceCatalog(sch)
    oneObjCatalog.table.defineCentroid(centroidName)
    
    s = oneObjCatalog.addNew()
    s.set("%s.x" % centroidName, x)
    s.set("%s.y" % centroidName, y)
    
    matches = [m.first for m in afwTable.matchXy(sourceSet, oneObjCatalog, radius)
               if m.first.get("deblend.nchild") == 0]

    if len(matches) == 0:
        raise RuntimeError("Unable to find a source within %g pixels of (%g,%g)" % (radius, x, y))

    if len(matches) == 1:
        return matches[0]
    else:
        return matches
```

```python
def addCcdVisitToOid(oid, visit, ccd):
    oid += (200*visit + ccd) << 32
    
    return oid

def splitExposureId(oid):
    """Split an ObjectId (maybe an numpy array) into visit, ccd, and objId.
See obs/subaru/python/lsst/obs/hscSim/hscMapper.py"""
    oid = np.array(oid, dtype='int64')
    objId = np.bitwise_and(oid, 2**32 - 1) # Should be the same value as was set by apps code
    oid = np.right_shift(oid, 32).astype('int32')

    ccd = (oid % 200).astype('int32')
    oid //= 200
    visit = oid.astype('int32')

    if visit.size == 1:     # sqlite doesn't like numpy types
        visit = int(visit)
        ccd = int(ccd)
        objId = int(objId)

    return dict(visit=visit, ccd=ccd, objId=objId)
```

```python
butler = dafPersist.Butler(os.path.join(os.environ["SUPRIME_DATA_DIR"], "rerun", "rhl", "cosmos"))

inDir = os.path.join(os.environ["SUPRIME_DATA_DIR"], "rerun", "rhl", "cosmos")
catTemplate="PF-%07d-%03d.fits"

cats = []
for visit in [908120, 908140, 908160]:
    if visit == 908120:
        continue
    for ccd in range(50, 59+1):
        cat = afwTable.SourceCatalog.readFits(os.path.join(inDir, catTemplate % (visit, ccd)))
        addCcdVisitToOid(cat.get("id"), visit, ccd)
        cats.append(cat)
        if False:
            exposure = butler.get("calexp", visit=visit, ccd=ccd)
        
pixelSize = afwImage.makeWcs(butler.get("calexp_md", visit=visit, ccd=ccd)).pixelScale().asArcseconds()
```

Concatenate all those individual catalogues

N.b. the centroids are still relative to the CCD origin

```python
cat = cats[0].copy(True)
cat.clear()        # Needed so that the reserve() call will generate contigous memory
cat.preallocate(sum([len(c) for c in cats]))

for c in cats:
    cat.extend(c, True)

print "%d objects" % (len(cat.get("id")))
```
